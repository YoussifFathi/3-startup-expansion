### Add Net_pofit and ROI Columns

```python
expansion_df['Net_Profit'] = expansion_df['Revenue'] - expansion_df['Marketing_Spend']
expansion_df
```

```python
expansion_df['ROI'] = ((expansion_df['Net_Profit'] / expansion_df['Marketing_Spend']) * 100).round(2)
expansion_df
```

```python
expansion_df[(expansion_df.New_Expansion == 'New') & (expansion_df.State == 'Tennessee')][['Marketing_Spend','Revenue','ROI']].mean()
```

```python
expansion_df[(expansion_df.New_Expansion == 'Old') & (expansion_df.State == 'Tennessee')][['Marketing_Spend','Revenue','ROI']].mean()
```

```python
gb_new_states = expansion_df[(expansion_df.New_Expansion == 'New')].groupby('State')[['Marketing_Spend','Revenue','ROI']].mean()
gb_new_states
```

```python
gb_old_states = expansion_df[(expansion_df.New_Expansion == 'Old')].groupby('State')[['Marketing_Spend','Revenue','ROI']].mean()
gb_old_states[gb_old_states.index.isin(gb_new_states.index)]
```

```python
merged_states_old_and_new = gb_new_states.merge(gb_old_states,how='inner',on='State',suffixes=('_new','_old'))
merged_states_old_and_new
```

```python
merged_states_net_new_old = pd.DataFrame()
merged_states_net_new_old['Net_MS'] = merged_states_old_and_new['Marketing_Spend_new'] - merged_states_old_and_new['Marketing_Spend_old']
merged_states_net_new_old['Net_Revenue'] = merged_states_old_and_new['Revenue_new'] - merged_states_old_and_new['Revenue_old']
merged_states_net_new_old['Net_ROI'] = merged_states_old_and_new['ROI_new'] - merged_states_old_and_new['ROI_old']
merged_states_net_new_old
```

### A3 : About City we did not open any new store in already old city
### About State : there State new stores doing better than old store in same states (Check Prev Table For More Statistical Info)

### based on Net_ROI : we see new stores in following state is doing better than old stores in the same state

```python
merged_states_net_new_old[merged_states_net_new_old.Net_ROI >= 0].index
```

### based on Net_ROI : we see new stores in following state is Not doing better than old stores in the same state

```python
merged_states_net_new_old[merged_states_net_new_old.Net_ROI < 0].index
```
